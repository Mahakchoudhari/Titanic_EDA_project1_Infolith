```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

df = pd.read_csv("train.csv")
df.head()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df.drop("Cabin", axis=1, inplace=True)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Survived", data=df, palette="Set2")
plt.title("Survival Count on Titanic")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

Takeaway: More passengers died than survived in the Titanic disaster.

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Sex", hue="Survived", data=df, palette="Set1")
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

Takeaway: Females had significantly higher survival rates than males.

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Pclass", hue="Survived", data=df, palette="Set3")
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

Takeaway: First-class passengers had the highest survival probability.

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=30, kde=True, color="blue")
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Takeaway: Most passengers were young adults between 20–40 years.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Survived", y="Fare", data=df, palette="coolwarm")
plt.title("Fare vs Survival")
plt.xlabel("Survived")
plt.ylabel("Fare")
plt.show()
```

Takeaway: Higher fare passengers had better survival chances.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

Takeaway: Passenger class and fare show strongest relationship with survival.

## Summary

The Titanic dataset reveals strong patterns behind passenger survival. Gender was the most important factor, with females showing significantly higher survival rates compared to males. Passenger class also played a major role, as first-class passengers had much better survival chances than those in third class, highlighting the impact of socioeconomic status.

Age distribution shows that most passengers were young adults, but children had slightly better survival outcomes compared to older individuals. Fare analysis further supports the class-based survival pattern, as higher-paying passengers were more likely to survive.

The correlation heatmap confirms that passenger class and fare are strongly linked with survival, while other variables like age show weaker influence.

Overall, survival on the Titanic was not random. It was strongly influenced by gender, wealth, and social class. This analysis demonstrates how data visualization can uncover meaningful patterns and tell a clear story from raw data.
